```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
import seaborn as sns
import csv
```

```python
df_house = pd.read_csv('kc_house_data.csv')
```

# read data

```python
df_house.head()
```

# display dataframe

```python
df_house.isnull().sum()
```

# this line checks for empty or missing data, this dataset is pretty clean with no missing values

```python
df_house.shape
```

```python
df_house = df_house[ (df_house['zipcode'] == 98039) ]
```

# select houses in a certain zip code

```python
y = df_house['price'].values
x = df_house['sqft_living'].values
```

# get data of price and square feet

```python
X = x.reshape(-1,1)
```

```python
x_train, x_test, y_train, y_test = train_test_split(X,y, train_size=0.8, test_size=0.2, random_state=100)
```

# here, we split the data into a training set and a testing set

```python
plt.scatter(x, y)
plt.xlabel('footage')
plt.ylabel('price')
plt.title('house sale data')
plt.show()
```

# then we plot our data with a scatter

```python
model = LinearRegression()
model.fit(x_train, y_train)
y_predict = model.predict(x_test)
print(f'train acc {round(model.score(x_train, y_train)*100,2)}%')
print(f'test acc {round(model.score(x_test, y_test)*100,2)}%')
```

# the accuracy of our test and training models is pretty high

```python
plt.scatter(x_train, y_train, color ='red')
plt.plot(x_test,y_predict)
plt.xlabel('footage')
plt.ylabel('price')
plt.title('house sale data')
plt.plot
```
